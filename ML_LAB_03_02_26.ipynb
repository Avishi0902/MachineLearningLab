CO-RELATION

In linear regression, one variable needs to depend on the other.
Values need to be continous.

```python
import numpy as np

x=np.array([1,3,5,7,8,9,10,15])
y=np.array([10,20,30,40,50,60,70,80])

def Coorelation(X,Y):
  if len(X)==len(Y):
    sum_xy = sum((X-X.mean())*(Y-Y.mean()))
    sum_x_squared= sum((X-X.mean())**2)
    sum_y_squared= sum((Y-Y.mean())**2)
    corr= sum_xy/np.sqrt(sum_x_squared*sum_y_squared)
  return corr
```

```python
print(Coorelation(x,y))
```

```python
print(Coorelation(x,x))
```

```python
print(Coorelation(y,y))
```

```python
print(Coorelation(y,x))
```
